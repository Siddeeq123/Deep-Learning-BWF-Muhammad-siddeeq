# Data Transformation

## Removing Duplicates


```python
data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two']*2,'k2': [1, 1, 2, 3, 3, 4, 4,4]})
data['v1'] = range(8)
data
```

```python
data.duplicated()
```

```python
data.drop_duplicates()
```

```python
data.drop_duplicates(["k2"])
```

```python
data.drop_duplicates(["k1"])
```

```python
data.drop_duplicates(['k1', 'k2'], keep='last')
```

## Transforming Data Using a Function or Mapping


```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon','Pastrami', 'corned beef', 'Bacon','pastrami', 'honey ham', 'nova lox'],'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python

meat_to_animal = {
 'bacon': 'pig',
 'pulled pork': 'pig',
 'pastrami': 'cow',
 'corned beef': 'cow',
 'honey ham': 'pig',
 'nova lox': 'salmon'
}
```

```python
lowercased = data['food'].str.lower()
```

```python
lowercased
```

```python
data["newcol"]=lowercased.map(meat_to_animal)
data
```

```python
data["food"].map(lambda x: meat_to_animal[x.lower()])
```

```python
data
```

```python
data = pd.Series([1., -999., 2., -999., -1000., 3.])
data.replace([-1000,-999],[np.nan,0])
```

```python
data = pd.DataFrame(np.arange(12).reshape((3, 4)), index=['Ohio', 'Colorado', 'New York'], columns=['one', 'two', 'three', 'four'])
data
```

```python
transform = lambda x: x[:4].upper()
```

```python
data.columns.map(transform)
```

```python
data.columns=data.columns.map(transform)
data
```

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
ages
```

```python
bins=[18,25,35,60,100]
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
cut=pd.cut(ages,bins,labels=group_names)
cut.value_counts()
```

```python
data=np.random.randn(17)
val=pd.cut(data,2,precision=1)
pd.value_counts(val)
```

```python
data=np.random.randn(20000)
cut=pd.qcut(data,2
            )
pd.value_counts(cut)
```
